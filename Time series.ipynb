This data is evenly spaced and has a 1 year resolution. Most wavelet require evenly spaced data. You can use the previously discucessed methods to interpolate. Pyleoclim uses Weighted-Wavelet Z transform which was desinged for unevenly-spaced data, however this can take a long time to run.

```python
data_ps_s =  data_ps.standardize()
data_ps_s.plot()
```

```python
scal = data_ps_s.wavelet()
scal.plot()
```

```python
scal_sig = scal.signif_test(method='ar1asym')
scal_sig.plot()
```

```python
scal_sig2 = scal.signif_test(method='ar1sim', number = 1000)
scal_sig2.plot()
```

Now try this on your data. If you have **unevenly** spaced data either interp() the data before appling the wavelet() or use the wwz method: **data_ps.wavelet(method='wwz')**. WWZ should be preferred, but might take more time. continous wavelet + interpolation can be used to get a first overview.


```python
data_ps.wavelet(method='wwz')
```

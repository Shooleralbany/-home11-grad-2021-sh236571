```python
import matplotlib.pyplot as plt
import pandas as pd
import pyleoclim as pyleo
import numpy as np
```

```python
# 1. Reading the data from the TXT file
data_dir = '/content/drive/MyDrive/Time Series'
# Open the dataset assuming it's a CSV file, change to pd.read_table for TSV files
data = pd.read_csv(data_dir+'/Challisdata.txt', sep='\t')
dates = data['CE'].tolist()
values1 = data['Weightorganic'].tolist()
values2 = data['13C'].tolist()
values3 = data['15N'].tolist()
values4 = data['CN'].tolist()
print(values1)

# Removing NaN values
non_nan_data = data[['CE', 'Weightorganic', '13C', '15N', 'CN']].dropna()
dates_non_nan = non_nan_data['CE'].tolist()
values1_non_nan = non_nan_data['Weightorganic'].tolist()
values2_non_nan = non_nan_data['13C'].tolist()
values3_non_nan = non_nan_data['15N'].tolist()
values4_non_nan = non_nan_data['CN'].tolist()

# 2. Create subplots
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 8))

axes[0].plot(values1_non_nan, dates_non_nan, label='Weight % Organic', linestyle='-', linewidth=1, alpha=0.7, color='black', marker='o')
axes[1].plot(values2_non_nan, dates_non_nan, label='δ13C(% VPDB)', linestyle='-', linewidth=1, alpha=0.7, color='black', marker='o')
axes[2].plot(values3_non_nan, dates_non_nan, label='δ15N(% AIR)', linestyle='-', linewidth=1, alpha=0.7, color='black', marker='o')
axes[3].plot(values4_non_nan, dates_non_nan, label='C/N', linestyle='-', linewidth=1, alpha=0.7, color='black', marker='o')


# 4. Customizing the subplots
axes[0].set_title('Weight % Organic')
axes[0].set_xlabel('')
axes[0].set_ylabel('Years (CE)')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(False)

axes[1].set_title('δ13C(% VPDB)')
axes[1].set_xlabel('')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(False)
axes[1].set_yticks([])
axes[1].spines['left'].set_visible(False)
axes[1].spines['bottom'].set_visible(False)
axes[1].spines['top'].set_visible(True)
axes[1].xaxis.tick_top()

axes[2].set_title('δ15N(% AIR)')
axes[2].set_xlabel('')
axes[2].set_yticks([])
axes[2].spines['left'].set_visible(False)
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(False)

axes[3].set_title('C/N')
axes[3].set_xlabel('')
axes[3].tick_params(axis='x', rotation=45)
axes[3].grid(False)
axes[3].set_yticks([])
axes[3].spines['left'].set_visible(False)
axes[3].spines['bottom'].set_visible(False)
axes[3].spines['top'].set_visible(True)
axes[3].xaxis.tick_top()

axes[0].set_ylim([-128, 2022])
axes[1].set_ylim([-128, 2022])
axes[2].set_ylim([-128, 2022])
axes[3].set_ylim([-128, 2022])

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
